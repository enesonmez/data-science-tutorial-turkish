#### Basit Bir Grafik Ã‡izimi

```python
ageList = [50,85,63,12,95,21,32,55,5,45]
weightList = [86,60,78,40,61,78,88,81,15,65]

ageArray = np.sort(np.array(ageList))
weightArray = np.sort(np.array(weightList))
```

```python
plt.plot(ageArray,weightArray,'g')
plt.xlabel('yas')
plt.ylabel('kilo')
plt.title("Kilo'nun YaÅŸa GÃ¶re DeÄŸiÅŸim GrafiÄŸi")
plt.show()
```

```python
array1 = np.linspace(0,10,20)
array2 = array1 ** 3
```

#### Figure (Ã–zelleÅŸirme)

Plt nesnesi bize belli yÃ¼kseklikte, belli geniÅŸlikte ve belli gÃ¶rÃ¼nÃ¼m Ã¶zelliklerine sahip bir nesne dÃ¶ndÃ¼rÃ¼r. Ancak biz kendimize Ã¶zgÃ¼ bir gÃ¶rselleÅŸtirme yapÄ±sÄ± kullanabiliriz. Bunun iÃ§inde **figure** adÄ±nÄ± vermiÅŸ olduÄŸumuz yapÄ±dan nesne Ã¼reterek kendi ayralamalarÄ±mÄ±zÄ± yapabiliriz.

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.6,0.6]) # [sol_eksen,saÄŸ_eksen,yÃ¼kseklik,geniÅŸlik]
figureAxes.plot(array1,array2,'g')
figureAxes.set_xlabel('x ekseni')
figureAxes.set_ylabel('y ekseni')
figureAxes.set_title('baÅŸlÄ±k')
```

#### Nested Graph (Ä°Ã§ iÃ§e Grafikler)

**add_axes** deÄŸerleri ile oynayarak tam olarak ne ifade ettiÄŸini anlayabilirsiniz. GrafiÄŸin konumunu deÄŸiÅŸtirebilmemizi saÄŸlamaktadÄ±r.

```python
figure1 = plt.figure()

axes1 = figure1.add_axes([0.1,0.1,0.9,0.9])
axes2 = figure1.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(array1,array2,'g')
axes1.set_xlabel('x ekseni')
axes1.set_ylabel('y ekseni')
axes1.set_title('ana baÅŸlÄ±k')

axes2.plot(array2,array1,'r')
axes2.set_xlabel('x ekseni')
axes2.set_ylabel('y ekseni')
axes2.set_title('kÃ¼Ã§Ã¼k baÅŸlÄ±k')
```

#### Subplots

**Subplots** geriye eksen ve figÃ¼r dÃ¶ndÃ¼rÃ¼r.

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot()
```

```python
type(myAxes)
```

```python
(myFigure2, myAxeses) = plt.subplots(nrows=1,ncols=2)
```

subplot'lar oluÅŸturunca geriye dÃ¶nen eksen'ler numpy array formatÄ±nda return edilirler.

```python
type(myAxeses)
```

```python
for axes in myAxeses:
    axes.plot(array1,array2,'g')
    axes.set_xlabel('x ekseni')


myFigure2.tight_layout() # EÄŸer grafikler iÃ§ iÃ§e geÃ§erse bu fonksiiyon ile aralarÄ±nÄ± aÃ§abilirsiniz.
myFigure2
```

```python
fig, axs = plt.subplots(3,3)
axs[0,0].set_facecolor('red')
axs[2,2].set_facecolor('yellow')
```

#### Legend KullanÄ±mÄ±

**figure parametreleri** <br>
**dpi** : gÃ¶rÃ¼ntÃ¼ kalitesini ayarlar <br>
**figuresize** : gÃ¶rÃ¼ntÃ¼nÃ¼n yÃ¼ksekliÄŸini ve geniÅŸliÄŸini ayarlar. tuple girdisi verilmeli.

```python
newFigure = plt.figure(dpi=100, figsize=(4,4))
newAxes = newFigure.add_axes([0.1,0.1,0.9,0.9])

newAxes.plot(array1, array1 ** 2, label='x**2')
newAxes.plot(array1, array2, label='x**3')
newAxes.legend(loc=6) ## loc ile legend'Ä±n grafik Ã¼zerindeki konumunu ayarlar.
```
