#### Basit Bir Grafik Çizimi

```python
ageList = [50,85,63,12,95,21,32,55,5,45]
weightList = [86,60,78,40,61,78,88,81,15,65]

ageArray = np.sort(np.array(ageList))
weightArray = np.sort(np.array(weightList))
```

```python
plt.plot(ageArray,weightArray,'g')
plt.xlabel('yas')
plt.ylabel('kilo')
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği")
plt.show()
```

```python
array1 = np.linspace(0,10,20)
array2 = array1 ** 3
```

```python
plt.plot(array1,array2,'b*-')
```

#### Subplot

```python
plt.subplot(1,2,1)
plt.plot(array1,array2,'r*-')
plt.subplot(1,2,2)
plt.plot(array2,array1,'g--')
```

#### Figure (Özelleşirme)

Plt nesnesi bize belli yükseklikte, belli genişlikte ve belli görünüm özelliklerine sahip bir nesne döndürür. Ancak biz kendimize özgü bir görselleştirme yapısı kullanabiliriz. Bunun içinde **figure** adını vermiş olduğumuz yapıdan nesne üreterek kendi ayralamalarımızı yapabiliriz.

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.6,0.6]) # [sol_eksen,sağ_eksen,yükseklik,genişlik]
figureAxes.plot(array1,array2,'g')
figureAxes.set_xlabel('x ekseni')
figureAxes.set_ylabel('y ekseni')
figureAxes.set_title('başlık')
```

#### Nested Graph (İç içe Grafikler)

**add_axes** değerleri ile oynayarak tam olarak ne ifade ettiğini anlayabilirsiniz. Grafiğin konumunu değiştirebilmemizi sağlamaktadır.

```python
figure1 = plt.figure()

axes1 = figure1.add_axes([0.1,0.1,0.9,0.9])
axes2 = figure1.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(array1,array2,'g')
axes1.set_xlabel('x ekseni')
axes1.set_ylabel('y ekseni')
axes1.set_title('ana başlık')

axes2.plot(array2,array1,'r')
axes2.set_xlabel('x ekseni')
axes2.set_ylabel('y ekseni')
axes2.set_title('küçük başlık')
```

#### Subplots

**Subplots** geriye eksen ve figür döndürür.

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot()
```

```python
type(myAxes)
```

```python
(myFigure2, myAxeses) = plt.subplots(nrows=1,ncols=2)
```

subplot'lar oluşturunca geriye dönen eksen'ler numpy array formatında return edilirler.

```python
type(myAxeses)
```

```python
for axes in myAxeses:
    axes.plot(array1,array2,'g')
    axes.set_xlabel('x ekseni')


myFigure2.tight_layout() # Eğer grafikler iç içe geçerse bu fonksiiyon ile aralarını açabilirsiniz.
myFigure2
```

```python
fig, axs = plt.subplots(3,3)
axs[0,0].set_facecolor('red')
axs[2,2].set_facecolor('yellow')
```

#### Legend Kullanımı

**figure parametreleri** <br>
**dpi** : görüntü kalitesini ayarlar <br>
**figuresize** : görüntünün yüksekliğini ve genişliğini ayarlar. tuple girdisi verilmeli.

```python
newFigure = plt.figure(dpi=100, figsize=(4,4))
newAxes = newFigure.add_axes([0.1,0.1,0.9,0.9])

newAxes.plot(array1, array1 ** 2, label='x**2')
newAxes.plot(array1, array2, label='x**3')
newAxes.legend(loc=6) ## loc ile legend'ın grafik üzerindeki konumunu ayarlar.
```
