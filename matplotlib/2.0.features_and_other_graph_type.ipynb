```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
array1 = np.linspace(0,10,20)
array2 = array1 ** 2
```

**color** : plot'un rengini belirler. <br> 
**alpha** : plot'un saydamlığını belirtir.

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot(array1, array2, color="#3eb289", alpha=0.5)
```

**linewidth** : çizginin kalınlığını değiştirir. <br>
**linestyle** : çizginin sitilini değiştirir. <br>
**marker** : değerlerin kesişim noktalarını belirtmek için kullanılır. (+,*,o,vb.) <br>
**markersize** : marker'ın boyutunu belirtir. <br>
**markerfacecolor** marker'ın rengini belirtir. <br>

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot(array1, array1 + 2, color="blue", linewidth=3.0, linestyle='--')
myAxes.plot(array1, array1 + 3, color="yellow", linewidth=5.0, linestyle=':')
myAxes.plot(array1, array1 + 4, color="red", linewidth=2.0, linestyle='-.')

myAxes.plot(array1, array1 + 5, color="black", linewidth=2.0, linestyle='--', marker='o', markersize=5, markerfacecolor='red')
```

### Scatter Plot

```python
plt.scatter(array1, array2, color = 'hotpink')
```

### Bar Plot

```python
plt.bar(array1, array2, width=0.1)
```

### Pie Plot

```python
z = np.random.uniform(0,1,4)
```

```python
myexplode = [0, 0.2, 0, 0]

plt.pie(z, labels=['x','y','z','l'], startangle=90, shadow=True, explode=myexplode)
plt.show()
```

### Grafiklerde Geçişli Renk Kullanımı

```python
x = np.random.randn(1000,4)
cmap = plt.get_cmap('Oranges')
colors = cmap([0.2,0.4,0.6,0.8])

plt.hist(x,color=colors)
```

### Grid Kullanımı

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid(axis='y', color='green', linestyle='--')
plt.show()
```
